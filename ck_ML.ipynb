```python
import tensorflow
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error
```

```python
diabetes = datasets.load_diabetes()
```

```python
X, Y = diabetes.data, diabetes.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.1)
```

```python
x_train.shape, y_train.shape
```

```python
linear = linear_model.LinearRegression()
linear.fit(x_train, y_train)
```

```python
linear.coef_
```

```python
linear.score(x_test, y_test)
```

```python
y_pre = linear.predict(x_test)
```

```python
mse = round(mean_squared_error(y_test, y_pre))
mse
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pre )
```

```python
alphas = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
mse_r = []
for alpha in alphas:
  r_diabetes = Ridge(alpha=alpha)
  r_diabetes.fit(x_train, y_train)
  y_pre_r = r_diabetes.predict(x_test)
  mse_r.append(round(mean_squared_error(y_test, y_pre_r),2))
print(*mse_r)
```

```python
alphas = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
mse_l = []
for alpha in alphas:
  l_diabetes = Lasso(alpha=alpha)
  l_diabetes.fit(x_train, y_train)
  y_pre_l = l_diabetes.predict(x_test)
  mse_l.append(round(mean_squared_error(y_test, y_pre_l),2))
print(*mse_l)
```
